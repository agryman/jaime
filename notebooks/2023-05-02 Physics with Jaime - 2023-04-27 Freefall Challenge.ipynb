## Problem Statement

Given:

* The elevator is at rest on the ground at time 0.
* The elevator has mass 1600 kg.
* The upward force exerted by the cable on the elevator was 27000 N.
* The elevator moved 3 m before the cable broke.

Question:

* With what velocity did the elevator hit the floor?

## Solution

First we compute the upward motion,
then we compute the downward motion.

### Upward Motion

Let's assume that the cable exerted a constant upward force up until it broke.

1 N = 1 kg$\cdot$m / s<sup>2</sup>

Let $T$ = 27000 N be the upward tension force.

Let $m$ = 1600 kg be the mass.

Let $G$ be the force due to gravity.
Note that $G$ is downward.

$$
G = -mg
$$

Let $F$ be the net force acting on the elevator.

\begin{align}
F &= T + G \\
&= T - mg
\end{align}

Newton's second law of motion is

\begin{align}
F &= m a \\
&= T - m g
\end{align}

Therefore the acceleration is

\begin{align}
a &= F / m \\
&= T / m - g
\end{align}

```python
T = 27000
m = 1600
g = 9.8

a = T / m - g

a
```

Let $t$ be the time travelled.

Let $v(t)$ be the upward velocity of the elevator at time $t$. 

$$
v(t) = a t
$$

Let $s(t)$ be the distance travelled at time $t$.

$$
s(t) = \frac{1}{2} a t^2
$$

Let $t_1$ be the time at which the cable broke.

Let $s_1$ be the height of the elevator above the floor at $t_1$.
We are told that the elevator moved 3 m at $t_1$.

$$
s_1 = s(t_1) = 3
$$

```python
s1 = 3
```

We can solve for $t_1$.

$$
s_1 = \frac{1}{2} a t_1^2
$$

$$
t_1 = \sqrt{\frac{2 s_1}{a}}
$$

```python
import math

t1 = math.sqrt(2 * s1 / a)

t1
```

Let $v_1$ be the upward velocity of the elevator at $t_1$.

$$
v_1 = v(t_1) = a t_1
$$ 

```python
v1 = a * t1

v1
```

### Downward Motion

When the cable breaks, the elevator falls under the influence of gravity.
Note that the elevator actually goes upward for a short time after the cable breaks, and then falls downward.

Let $g$ be the acceleration due to gravity.

$g = 9.8$ N

```python
g = 9.8
```

For times $t$ greater than $t_1$, the velocity is

$$
v(t) = v_1 - g (t - t_1)
$$

The height above the floor for $t \ge t_1$ is

$$
s(t) = s_1 + v_1 (t - t_1) - \frac{1}{2} g (t - t_1)^2
$$

Let $t_2$ be the time at which the elevator hits the floor.

Then the height is zero at $t_2$.

$$
s(t_2) = 0
$$

We can solve for $t_2$.

$$
0 = s_1 + v_1 (t_2 - t_1) - \frac{1}{2} g (t_2 - t_1)^2
$$

Let $x$ be the difference between $t_2$ and $t_1$.

$$
x = t_2 - t_1
$$

Therefore

$$
-\frac{1}{2} g x^2 + v_1 x + s_1 = 0
$$

This is a quadratic equation. It's two solutions are

$$
x = \frac{-v_1 \pm \sqrt{v_1^2 + 2 g s_1}}{-g}
$$

or

$$
x = \frac{v_1 \mp \sqrt{v_1^2 + 2 g s_1}}{g}
$$

Since $t_2 \gt t_1$ we know that $x \gt 0$. 
The solution of interest is

$$
x = \frac{v_1 + \sqrt{v_1^2 + 2 g s_1}}{g}
$$

```python
x = (v1 + math.sqrt(v1 ** 2 + 2 * g * s1)) / g

x
```

Let $v_2$ be the velocity at $t_2$.

$$
v_2 = v(t_2) = v_1 - g (t_2 - t_1) = v_1 - g x
$$

```python
v2 = v1 - g * x

v2
```
